```python
x = dataset.iloc[:,1:31].values
y = dataset.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority',random_state=0)
x_SMOTE, y_SMOTE = smote.fit_resample(x_train, y_train)
```

```python
from imblearn.over_sampling import KMeansSMOTE
kmeanssmote = KMeansSMOTE(sampling_strategy='minority',random_state=0)
x_KMeansSMOTE, y_KMeansSMOTE = kmeanssmote.fit_resample(x_train, y_train)
```

```python
from imblearn.combine import SMOTETomek
smotetomek = SMOTETomek(sampling_strategy='minority',random_state=0)
x_SMOTETomek, y_SMOTETomek = smotetomek.fit_resample(x_train, y_train)
```

```python
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logistic_model = LogisticRegression(random_state = 0, solver='newton-cg')
logistic_model.fit(x_SMOTETomek, y_SMOTETomek)

y_predict = logistic_model.predict(x_test)

fpr, tpr, _ = metrics.roc_curve(y_test,  y_predict)
auc = metrics.roc_auc_score(y_test, y_predict)
result_table = result_table.append({'classifiers':'Logistic Regression',
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)

cf_matrix = metrics.confusion_matrix(y_test,y_predict)
group_names = ['True -ve', 'False +ve','False -ve','True +ve']
group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')

accuracy = (cf_matrix[0,0] + cf_matrix[1,1]) / sum(sum(cf_matrix))
print('Accuracy : ', accuracy)

sensitivity = cf_matrix[1,1] / (cf_matrix[1,1] + cf_matrix[1,0])
print('Sensitivity : ', sensitivity)
print('False Omission Rate : ', (1-sensitivity))

specificity = cf_matrix[0,0] / (cf_matrix[0,0] + cf_matrix[0,1])
print('Specificity : ', specificity)
print('False Discovery Rate : ', (1-specificity))

print('Matthews Correlation Coefficient : ', metrics.matthews_corrcoef(y_test, y_predict))
```

```python
output=pd.DataFrame(data={"Actual":y_test,"Prediction":y_predict})
output.to_csv(path_or_buf="Data/lr.csv",index=False,quoting=3,sep=',')
```

## K-nearest Neighbour Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

Ks = 25
mean_acc = np.zeros((Ks-1))
Y_PREDICT = []

for n in range(1,Ks): 
    knn_model = KNeighborsClassifier(n_neighbors = n).fit(x_KMeansSMOTE,y_KMeansSMOTE)
    y_predict = knn_model.predict(x_test)
    Y_PREDICT.append(y_predict)
    mean_acc[n-1] = metrics.accuracy_score(y_test, y_predict)
    
fpr, tpr, _ = metrics.roc_curve(y_test,  Y_PREDICT[mean_acc.argmax()])
auc = metrics.roc_auc_score(y_test, Y_PREDICT[mean_acc.argmax()])
result_table = result_table.append({'classifiers':'K-Nearest Neighbour',
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)

cf_matrix = metrics.confusion_matrix(y_test, Y_PREDICT[mean_acc.argmax()])
group_names = ['True -ve', 'False +ve','False -ve','True +ve']
group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')

accuracy = (cf_matrix[0,0] + cf_matrix[1,1]) / sum(sum(cf_matrix))
print("The best accuracy was", accuracy, "with k=", mean_acc.argmax()+1)

sensitivity = cf_matrix[1,1] / (cf_matrix[1,1] + cf_matrix[1,0])
print('Sensitivity : ', sensitivity)
print('False Omission Rate : ', (1-sensitivity))

specificity = cf_matrix[0,0] / (cf_matrix[0,0] + cf_matrix[0,1])
print('Specificity : ', specificity)
print('False Discovery Rate : ', (1-specificity))

print('Matthews Correlation Coefficient : ', metrics.matthews_corrcoef(y_test, y_predict))
```

```python
output=pd.DataFrame(data={"Actual":y_test,"Prediction":y_predict})
output.to_csv(path_or_buf="Data/knn.csv",index=False,quoting=3,sep=',')
```
